```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("canada_per_capita_income.csv")
df
```

```python
df.columns
```

In this `canada_per_capita_income.csv` file two columns `year` and `per capita income(US$)`, so I don't want to write this long string column name everytime so I rename it using `df.rename(columns={'per capita income (US$)': 'income'}, inplace=True)`

```python
df.rename(columns={'per capita income (US$)': 'income'}, inplace=True)
df.columns
```

We also rename using `df.columns = ['year', 'income']` but In this way we need to rename all the available columns.


```python
%matplotlib inline
plt.xlabel('Year')
plt.ylabel('income')
plt.scatter(df.year, df.income)
```

```python
year_df = df.drop('income', axis="columns")
year_df
```

```python
model = linear_model.LinearRegression()
model.fit(year_df, df.income)
```

```python
model.predict([[2020]])
```
